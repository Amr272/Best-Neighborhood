    The project was to find the best neighborhood using some data set to back up the conclusion. I wasn't sure what would make the best neighborhood, there is so many diffrent variables and things to consider, but I came to the conclusion that the Neighborehood with the least crime would be bestter than a place with more crime. I first looked at the crime data set and it had a massive amount of data and went back to 2005. I felt that including that much data might make areas that now are fine seem worse. To combat this I decided to use the monthly data set of police reports to have more up to date data. The data set that I used was "Police Incident Blotter (30 Day)".

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
# crime data
crimeData= pd.read_csv("CrimeDataplot.csv")

#output incident totals
crimeData['INCIDENTNEIGHBORHOOD'].value_counts()
```

```python
crimeData['INCIDENTNEIGHBORHOOD'].value_counts().plot.barh(figsize=(15,20), stacked =True)

plt.title("Crime by Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("# of Incidents")
```
